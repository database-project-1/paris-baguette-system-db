```python
customers = []
members = []
nonmembers = []
consents = []
consent_items = ['마케팅수신동의', '위치정보동의', '제3자정보제공동의']
used_usernames = set()

for i in range(1, 201):
    cid = f'C{i:04d}'
    name = fake.name()
    customers.append((cid, name))

    if i <= 150:  # 150명 회원
        # 중복 없는 아이디 생성
        while True:
            username = fake.user_name()
            if username not in used_usernames:
                used_usernames.add(username)
                break
        members.append(( #순서대로 만들기
            cid,
            username,
            f'pw{i:04d}',                                      # 비밀번호
            fake.phone_number(),                               # 연락처
            fake.email(),                                      # 이메일
            fake.address(),                                    # 주소
            fake.date_between('-3y','today').isoformat(),      # 가입일자
            random.randint(0,1),                               # 정보제공동의여부
        ))
        for item in consent_items:
            agreed = random.randint(0,1)
            consents.append((
                cid, item, fake.date_between('-2y','today').isoformat(),
                agreed, fake.date_between('today','+1y').isoformat() if agreed==0 else None
            ))
    else:  # 50명 비회원
        nonmembers.append((cid, random.randint(0,1)))

cursor.executemany("INSERT INTO 고객 VALUES (?,?)", customers)
cursor.executemany("INSERT INTO 회원 VALUES (?,?,?,?,?,?,?,?)", members)  # 컬럼 8개
cursor.executemany("INSERT INTO 비회원 VALUES (?,?)", nonmembers)
cursor.executemany("INSERT INTO 동의내역 VALUES (?,?,?,?,?)", consents)
conn.commit()
print(f"고객 {len(customers)}명 (회원 {len(members)}, 비회원 {len(nonmembers)}) 완료!")
```

```python
all_product_ids = [p[0] for p in products]
inventory = []
for branch in branch_names:
    # 각 지점마다 상품 중 랜덤하게 50~70개 보유
    held = random.sample(all_product_ids, random.randint(50, 70))
    for pid in held:
        price = next(p[2] for p in products if p[0]==pid)
        inventory.append((
            branch, pid,
            random.randint(0, 100),
            random.randint(5, 20),
            int(price * random.uniform(0.95, 1.1))
        ))

cursor.executemany("INSERT INTO 보유 VALUES (?,?,?,?,?)", inventory)
conn.commit()
print(f"재고 {len(inventory)}건 완료!")
```

```python
from datetime import date, timedelta

employee_ids = [e[0] for e in employees]
customer_ids = [c[0] for c in customers]
branch_emp = {}
for e in employees:
    branch_emp.setdefault(e[5], []).append(e[0])

sales = []
sale_details = []
payment_methods = ['현금', '카드', '간편결제']

for i in range(1, 1201):
    sid = f'S{i:05d}'
    branch = random.choice(branch_names)
    emp = random.choice(branch_emp[branch])
    cid = random.choice(customer_ids)
    sale_date = fake.date_between(start_date=date(2025, 12, 1), end_date=date(2026, 6, 16)).isoformat()

    # 해당 지점 보유 상품만 판매
    branch_products = [inv[1] for inv in inventory if inv[0]==branch]
    if not branch_products:
        continue

    items = random.sample(branch_products, min(random.randint(1,5), len(branch_products)))
    total = 0
    for j, pid in enumerate(items, 1):
        qty = random.randint(1, 5)
        price = next(p[2] for p in products if p[0]==pid)
        total += price * qty
        sale_details.append((sid, j, qty, pid, price))

    sales.append((sid, sale_date, total, random.choice(payment_methods), emp, cid, branch))

cursor.executemany("INSERT INTO 판매 VALUES (?,?,?,?,?,?,?)", sales)
cursor.executemany("INSERT INTO 판매상세 VALUES (?,?,?,?,?)", sale_details)
conn.commit()
print(f"판매 {len(sales)}건, 판매상세 {len(sale_details)}건 완료!")
```

```python
#fk 연결확인
checks = [
    ("상품 브랜드 연결", "SELECT COUNT(*) FROM 상품 WHERE 브랜드코드 NOT IN (SELECT 브랜드코드 FROM 브랜드)"),
    ("직원 지점 연결", "SELECT COUNT(*) FROM 직원 WHERE 지점명 NOT IN (SELECT 지점명 FROM 지점)"),
    ("판매 직원 연결", "SELECT COUNT(*) FROM 판매 WHERE 직원ID NOT IN (SELECT 직원ID FROM 직원)"),
    ("판매 고객 연결", "SELECT COUNT(*) FROM 판매 WHERE 고객ID NOT IN (SELECT 고객ID FROM 고객)"),
    ("판매상세 상품 연결", "SELECT COUNT(*) FROM 판매상세 WHERE 상품코드 NOT IN (SELECT 상품코드 FROM 상품)"),
    ("회원 고객 연결", "SELECT COUNT(*) FROM 회원 WHERE 고객ID NOT IN (SELECT 고객ID FROM 고객)"),
    ("발주상세 상품 연결", "SELECT COUNT(*) FROM 발주상세 WHERE 상품코드 NOT IN (SELECT 상품코드 FROM 상품)"),
]

print("=== FK 정합성 검증 ===")
all_pass = True
for name, query in checks:
    cursor.execute(query)
    count = cursor.fetchone()[0]
    status = "OK" if count == 0 else f"오류 {count}건"
    if count > 0:
        all_pass = False
    print(f"{name}: {status}")

print("\n전체 결과:", "모두 통과" if all_pass else "오류 존재")
```

```python
#제약조건 확인
checks2 = [
    ("가격 음수 여부", "SELECT COUNT(*) FROM 상품 WHERE 가격 < 0"),
    ("급여 0 이하 여부", "SELECT COUNT(*) FROM 직원 WHERE 급여 <= 0"),
    ("판매수량 0 이하 여부", "SELECT COUNT(*) FROM 판매상세 WHERE 수량 <= 0"),
    ("재고수량 음수 여부", "SELECT COUNT(*) FROM 보유 WHERE 재고수량 < 0"),
    ("결제수단 유효성", "SELECT COUNT(*) FROM 판매 WHERE 결제수단 NOT IN ('현금','카드','간편결제')"),
    ("발주상태 유효성", "SELECT COUNT(*) FROM 발주 WHERE 상태 NOT IN ('대기','완료')"),
]

print("=== CHECK 제약조건 검증 ===")
all_pass = True
for name, query in checks2:
    cursor.execute(query)
    count = cursor.fetchone()[0]
    status = "OK" if count == 0 else f"오류 {count}건"
    if count > 0:
        all_pass = False
    print(f"{name}: {status}")

print("\n전체 결과:", "모두 통과" if all_pass else "오류 존재")
```
